```python
import json
import requests
import os
import pandas as pd
#Using the `Ticker` module we can create an object that will allow us to access functions to extract data. \
#To do this we need to provide the ticker symbol for the stock, here the company is Apple and the ticker symbol is `AAPL`.
```

```python
with open('jobs2.json') as json_file:
    job_list = json.load(json_file)
```

```python
count=0
for data in job_list:
    job_title = data.get('Job Title')
    if  (job_title.find('Python')!=-1) :
            count=count+1
           # print(data.get('Id'), job_title )
print ('Total:' , count)
    
   

```

```python
def get_number_of_jobs_T(technology):
  number_of_jobs=0
  for data in job_list:
    job_title = data.get('Job Title')
    job_id =  data.get('Id')
    if  (job_title.find(technology)!=-1) :
            number_of_jobs+=1
           # print("Job ID: ",job_id," - > ", job_title )
  return technology,number_of_jobs
```

```python
get_number_of_jobs_T("Python")
    
```

```python
 def get_number_of_jobs_L(location):
    number_of_jobs=0
    for data in job_list:
        job_location = data.get('Location')
        job_id =  data.get('Id')
        if  (job_location.find(location)!=-1) :
            number_of_jobs+=1
           # print("Job ID: ", job_id," - > ", job_location )
    return location,number_of_jobs
```

```python
 get_number_of_jobs_L("Los Angeles")
```

```python
Tech_list=["Python", "Adobe" , "Azure" , "Oracle","Java"]
```

```python
for job_tech in Tech_list:
   print(get_number_of_jobs_T(job_tech))
```

```python
from openpyxl import Workbook        # import Workbook class from module openpyxl
wb=Workbook()                        # create a workbook object
ws=wb.active  
for job_tech in Tech_list:

    ws.append(get_number_of_jobs_T(job_tech))   # add a row with two columns 'Country' and 'Continent'
wb.save("job-postings.xlsx")            # save the workbook into a file called countries.xlsx
```
